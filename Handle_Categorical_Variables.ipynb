2. One Hot Encoding with Multiple Categories

```python
data=pd.read_csv('mercendez.csv',usecols=['X1','X2','X3','X4','X5','X6'])
data.head()
```

```python
# Lets have a look at how many labels each variable has - 
for cols in data.columns:
    print(cols,': ',len(data[cols].unique()),' labels')
```

```python
# How many columns we will have after one-hot encoding

pd.get_dummies(data,drop_first=True).shape
```

```python
# Top 20 most frequncy categories for variable X2

data.X2.value_counts().sort_values(ascending=False).head(20)
```

```python
# A list of 10 most frequent categories of variable

top_10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]

top_10
```

```python
#We make 10 binary variables

for label in top_10:
    data[label]=np.where(data['X2']==label,1,0)

data[['X2']+top_10].head(40)
```
